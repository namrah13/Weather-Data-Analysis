# Exploratory Data Analysis 

```python
# cleaning the data
```

```python
data.isnull().sum()
```

```python
data_drop = data.drop(['Dew Point Temp_C'],axis=1)
```

```python
data_drop.head()
```

```python
# relationship analysis
```

```python
correlation = data_drop.corr()
```

```python
# Heatmap
sns.heatmap(correlation, xticklabels=correlation.columns, yticklabels=correlation.columns, annot=True)
```

```python
# Pairplot
sns.pairplot(data_drop)
```

```python
# relation plot
sns.relplot(x='Temp_C', y='Press_kPa',hue='Visibility_km', data=data_drop)
```

```python
# distribution plot
sns.distplot(data_drop['Press_kPa'], bins=5)
```

```python
#categorical plot
sns.catplot(x='Wind Speed_km/h',kind='box',data=data_drop)
```
